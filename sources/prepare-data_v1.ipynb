```python
file_1_original = orginal_datasets_path + 'UNSW_2018_IoT_Botnet_Full5pc_1.csv'
file_2_original = orginal_datasets_path + 'UNSW_2018_IoT_Botnet_Full5pc_2.csv'
file_3_original = orginal_datasets_path + 'UNSW_2018_IoT_Botnet_Full5pc_3.csv'
file_4_original = orginal_datasets_path + 'UNSW_2018_IoT_Botnet_Full5pc_4.csv'
```

```python
import pandas as pd
df = pd.read_csv (file_1_original)
df2 = pd.read_csv (file_2_original)
df3 = pd.read_csv (file_3_original)
df4 = pd.read_csv (file_4_original)


df = df.append(df2)
df = df.append(df3)
df = df.append(df4)

```

```python
df
```

```python
tmp_df = df
```

```python
tmp_df.head(4)
```

```python
tmp_df.drop('stime', axis=1, inplace=True)
tmp_df.drop('ltime', axis=1, inplace=True)
tmp_df.drop('pkSeqID', axis=1, inplace=True)
tmp_df.drop('saddr', axis=1, inplace=True)
tmp_df.drop('daddr', axis=1, inplace=True)
tmp_df.drop('sport', axis=1, inplace=True)
tmp_df.drop('dport', axis=1, inplace=True)
tmp_df.drop('seq', axis=1, inplace=True)
```

```python
tmp_df
```

```python
flgs_dummies = pd.get_dummies(tmp_df.flgs, prefix='flgs')
tmp_df = pd.concat([tmp_df, flgs_dummies], axis=1)


flgs_number_dummies = pd.get_dummies(tmp_df.flgs_number, prefix='flgs_number')
tmp_df = pd.concat([tmp_df, flgs_number_dummies], axis=1)

proto_dummies = pd.get_dummies(tmp_df.proto, prefix='proto')
tmp_df = pd.concat([tmp_df, proto_dummies], axis=1)

proto_number_dummies = pd.get_dummies(tmp_df.proto_number, prefix='proto_number')
tmp_df = pd.concat([tmp_df, proto_number_dummies], axis=1)

pkts_dummies = pd.get_dummies(tmp_df.pkts, prefix='pkts')
tmp_df = pd.concat([tmp_df, pkts_dummies], axis=1)

state_dummies = pd.get_dummies(tmp_df.state, prefix='state')
tmp_df = pd.concat([tmp_df, state_dummies], axis=1)

state_number_dummies = pd.get_dummies(tmp_df.state_number, prefix='state_number')
tmp_df = pd.concat([tmp_df, state_number_dummies], axis=1)


spkts_dummies = pd.get_dummies(tmp_df.spkts, prefix='spkts')
tmp_df = pd.concat([tmp_df, spkts_dummies], axis=1)

dpkts_dummies = pd.get_dummies(tmp_df.dpkts, prefix='dpkts')
tmp_df = pd.concat([tmp_df, dpkts_dummies], axis=1)

tmp_df.drop('flgs', axis=1, inplace=True)
tmp_df.drop('flgs_number', axis=1, inplace=True)
tmp_df.drop('proto', axis=1, inplace=True)
tmp_df.drop('proto_number', axis=1, inplace=True)
tmp_df.drop('pkts', axis=1, inplace=True)
tmp_df.drop('state', axis=1, inplace=True)
tmp_df.drop('state_number', axis=1, inplace=True)
tmp_df.drop('spkts', axis=1, inplace=True)
tmp_df.drop('dpkts', axis=1, inplace=True)

```
